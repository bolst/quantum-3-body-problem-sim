### Imports

```python
import numpy as np
import numba
import matplotlib.pyplot as plt
import scipy.fft
import scipy.linalg
from scipy.integrate import simpson as integrate
%config InlineBackend.figure_format = 'retina'
plt.rcParams["figure.figsize"] = (4,4)
```

In our case, we have defined our Hamiltonian for a Three-Body Harmonic Oscillator in the following way:
\begin{equation}
\hat{H} \equiv \hat{H}_1 + \hat{H}_2 + \hat{H}_{CM} = \frac{\hat{Q}_1^2}{2m} + \frac{3}{2}\kappa \hat{R}_1^2 + \frac{\hat{Q}_2^2}{2m} + \frac{3}{2}\kappa \hat{R}_2^2 + \frac{\hat{Q}_3^2}{2m}
\end{equation}

where

$$
\newcommand{wave}{\Psi(\mathbf{r}_1, \mathbf{r}_2, \mathbf{r}_3)}
\newcommand{potential}{V(\mathbf{r}_1, \mathbf{r}_2, \mathbf{r}_3)}
\renewcommand{\O}[1]{\mathbf{\hat{#1}}}
\braket{\hat{Q}_1, \hat{Q}_2, \hat{Q}_3} = \braket{\frac{1}{\sqrt{2}}(\O{p}_1 - \O{p}_2), \frac{1}{\sqrt{6}}(2\O{p}_3 - \O{p}_1 - \O{p}_2), \frac{1}{\sqrt{3}}(\O{p}_1 + \O{p}_2 + \O{p}_3)}
$$
$$
\braket{\hat{R}_1, \hat{R}_2, \hat{R}_3} = \braket{\frac{1}{\sqrt{2}}(\O{r}_1 - \O{r}_2), \frac{1}{\sqrt{6}}(2\O{r}_3 - \O{r}_1 - \O{r}_2), \frac{1}{\sqrt{3}}(\O{r}_1 + \O{r}_2 + \O{r}_3)}
$$

For convenience I will rewrite $ \hat{H} $ in the following form

\begin{equation}
\hat{H} = \left(\frac{\hat{Q}_1^2 + \hat{Q}_2^2 + \hat{Q}_3^2}{2m} + \frac{3}{2}\kappa \left(\hat{R}_1^2 + \hat{R}_2^2\right)\right) \equiv \left(\frac{\O{p}^2}{2m} + V\right)
\end{equation}

So that we can express the TDSE as

\begin{equation}
i\hbar \frac{\partial}{\partial t}\wave = \left(\frac{\O{p}^2}{2m} + \potential\right)\wave
\end{equation}

For our wavefunction, I will use an initial state of a Gaussian wave packet: $\wave = \exp\left[-\frac{1}{2}\left(\hat{R}_1^2 + \hat{R}_2^2 + \hat{R}_3^2\right)\right]$. I chose this because $\wave\equiv\Psi_1(R_1)\Psi_2(R_2)\Psi_3(R_3)$.

```python
κ = 0.05
ħ = 1
m = 1
L  = 20

# manually compute normalization constant of a distribution
def compute_norm(G,X):
    return 1 / np.sqrt(integrate(y=np.abs(G)**2, x=X))

# convert r to R
def R_r(r1,r2,r3):
    return [(r1-r2)/np.sqrt(2), (2*r3-r1-r2)/np.sqrt(6), (r1+r2+r3)/np.sqrt(3)]

# normalized Gaussian constructor
# returns 3d array of each component ( e^(r1+r2+r3) = (e^r1)(e^r2)(e^r3) )
def Ψ_r(r1,r2,r3,X):
    # Jacobi variables
    R1,R2,R3 = R_r(r1,r2,r3)
    R1 = X - R1
    R2 = X - R2
    R3 = X - R3

    retval = []
    for R in [R1,R2,R3]:
        G = np.exp(-0.5*R**2)
        C = compute_norm(G,X)
        retval.append( C * G )
    return retval

# potential wrt r
def V_r(r1,r2,r3,X=0,κ=κ):
    R1,R2,R3 = R_r(r1,r2,r3)
    return V_R(R1,R2,R3,X)

# potential wrt Jacobi coordinates
def V_R(R1,R2,R3,X=0,κ=κ):
    R1 = X - R1
    R2 = X - R2
    R3 = X - R3
    return (3/2)*κ*(R1**2 + R2**2)
```

We're solving the time-dependent Schrodinger equation (in 1D)
$$
i\hbar \frac{\partial \psi(r_1,r_2,r_3,t)}{\partial t} = 
\left(\frac{\O{p}^2}{2m} + V(r_1,r_2,r_3)\right) \psi(r_1,r_2,r_3,t)
\equiv \hat{H} \psi(r_1,r_2,r_3,t)
$$
with periodic boundary conditions on an interval $[-L/2,L/2]$. Recall that $\hat{p_\mu} \equiv -i\hbar \frac{d}{dr_\mu}$. We will use a split-step and approximate the time-evolution operator $\hat{U}(\delta_t)$ for small $\delta_t$ as
$$
\psi(r_1,r_2,r_3,t+\delta_t) =
e^{-i\delta_t \hat{H}/\hbar} \psi(r_1,r_2,r_3,t)
\approx 
e^{-i\delta_t \left(\hat{Q_1^2} + \hat{Q_2^2} + \hat{Q_3^2}\right)/(2m \hbar)}e^{-i\delta_t V(r_1,r_2,r_3)/\hbar} \psi(r_1,r_2,r_3,t)
$$

We can treat the momentum part of each body separately, since $ Q_1^2 + Q_2^2 + Q_3^2 = p_1^2 + p_2^2 + p_3^2 $. For the potential part, we need to know $r_1, r_2, r_3$ at each step. I will make a relatively bold assumption that the peak of each wave packet corresponds to its $r$.

We'll do this on a finite grid with points $r_{\mu,n} \equiv n \delta_{r_\mu}$ where $n=0,1,2,\cdots,N_x-1$ and $\delta_{r_\mu} = L/N_x$.

We will also leverage FFT's from `scipy.fft` for calculating our momentum steps

```python
def fft_fwd(ψ,L=L): # from ψ to ψ tilde
    Nx = len(ψ)
    ψt = 1.0/Nx*scipy.fft.fft(ψ*(-1)**np.arange(0,Nx))
    return ψt    

def fft_bwd(ψt,L=L): # ψ tilde -> ψ
    Nx = len(ψt)
    ψ = Nx*scipy.fft.ifft(ψt)*(-1)**np.arange(0,Nx)
    return ψ
```

```python
def potential_step(ψ0, # complex array of size Nx
                   V0, # real array of size Nx
                   X,  # real array of size Nx
                   δt, # time step
                   ħ=ħ):
    ψ = np.exp(-1j*V0*δt/ħ)*ψ0 # element-wise multiplication

    # update potential
    Nx = len(X)
    ψ01,ψ02,ψ03 = ψ0
    R1,R2,R3 = X[np.argmax(ψ01)],X[np.argmax(ψ02)],X[np.argmax(ψ03)] # assume R1,R2,R3 are the max of each distribution
    R1,R2,R3 = np.linspace(R1,R1,Nx), np.linspace(R2,R2,Nx), np.linspace(R3,R3,Nx) 
    V = np.full(Nx, V_R(R1,R2,R3,X))
    
    return ψ,V

def kinetic_step(ψ0, # complex array of size Nx
                 δt, # time step
                 ħ=ħ,L=L,m=m):
    Nx = len(ψ0)
    
    # to momentum space
    ψt1 = fft_fwd(ψ0,L=L)
    
    # evolved in momentum space
    K = 2.0*np.pi/L*np.arange(-Nx//2,Nx//2)
    E = ħ**2*K**2/(2*m) # energy of a wave packet = ħ^2 k^2 / 2m
    ψt2 = np.exp(-1j*δt/ħ*E)*ψt1
    
    # from momentum space    
    return fft_bwd(ψt2,L=L)
```

```python
# step method
def verlet_split(ψ0,V0,X,δt,ħ=ħ,m=m,L=L):    
    χ1,V1  = potential_step(ψ0,V0,X,0.5*δt,ħ=ħ)   

    # compute for r1,r2,r3
    χ2 = []
    for χ in χ1:
        χ2i    = kinetic_step(χ,δt,ħ=ħ,m=m,L=L)
        C      = compute_norm(χ2i,X)
        χ2.append(C * χ2i)
    
    χ3,V = potential_step(χ2,V1,X,0.5*δt,ħ=ħ)
    return χ3,V

# evolver
def evolve_split_step(ψ0,V0,X,δt,Nt,L=L,m=m,ħ=ħ,step=verlet_split):
    Nx = len(X)
    T = np.linspace(0,Nt*δt,Nt,endpoint=False)
    Ψ = np.zeros((3,Nx,Nt),dtype=np.complex128)
    V = np.zeros((Nx,Nt),dtype=np.complex128)
    
    Ψ[:,:,0] = ψ0
    V[:,0] = V0
    for it in range(Nt-1):
        Ψ[:,:,it+1],V[:,it+1] = step(Ψ[:,:,it],V[:,it],X,δt,ħ=ħ,m=m,L=L)
    return T,Ψ,np.real(V)
```

```python
Nx = 1000
X = np.linspace(-L/2,L/2,Nx,endpoint=False)

# initial conditions
r1,r2,r3 = 1,3,-4
r1,r2,r3 = np.linspace(r1,r1,Nx), np.linspace(r2,r2,Nx), np.linspace(r3,r3,Nx)
V0 = V_r(r1,r2,r3,X)
V = np.full(Nx,V0)

ψ01,ψ02,ψ03 = Ψ_r(r1,r2,r3,X)

tf  = 8*np.pi
Nt = 5000
δt = tf/Nt

plt.plot(X,V,color='gray')
plt.plot(X,np.real(ψ01),color='r')
plt.plot(X,np.imag(ψ01),color='r')
plt.plot(X,np.real(ψ02),color='b')
plt.plot(X,np.imag(ψ02),color='b')
plt.plot(X,np.real(ψ03),color='g')
plt.plot(X,np.imag(ψ03),color='g')
plt.ylim(-0.5,5)
#plt.xlim(-10,10)
plt.xlabel('x')
```

```python
%time T,Ψ,V = evolve_split_step([ψ01,ψ02,ψ03],V,X,δt,Nt,L=L)
```

```python
import ipywidgets as widgets
@widgets.interact(t=(0.0,tf,δt))
def disp(t=0):
    it = int(t/tf*(Nt-1))
    for i,color in zip(range(3),['r','b','g']):
        φ = V[:,it]
        ψ = Ψ[i,:,it]
        plt.plot(X,φ,color='gray')
        #plt.plot(X,np.real(ψ),color=color)
        #plt.plot(X,np.imag(ψ),color=color)
        plt.plot(X,np.abs(ψ),color=color)
    plt.ylim(-0.5,4)
    plt.xlim(-L/2,L/2)
```

### Energy Bounds

I will use a numerical diagonalization routine to solve a 1D time-independent Schrodinger equation. I will use it to find the first few energy levels and see if the lower bound holds. This TISE can be discretized and expressed as

\begin{equation}
-\frac{\hbar^2}{2m\delta_x^2}\left[\psi_{n+1}-2\psi_n+\psi_{n-1}\right] + V_n\psi_n = E_n\psi_n
\end{equation}

where $\psi_n\equiv\psi(x_n)$ and $x_n=n\delta_x$ and $\delta_x=L/N_x$. This gives a matrix eigenvalue equation in the form

\begin{equation}
\sum_{n'}H_{n,n'}\psi_{n'} = E_n\psi_n
\end{equation}

where the matrix $H_{n,n'}$ is given as

\begin{equation}
H_{n,n'} = \left(\begin{array}{ccccc}
2\alpha + V_1 & -\alpha & 0  & \cdots  & 0 \\
-\alpha & 2\alpha + V_2 & -\alpha & \cdots  & 0 \\
0 & -\alpha & 2\alpha + V_3 & \cdots  & 0 \\
\vdots & \vdots & \vdots & \ddots & \vdots \\
0 & 0 & 0 & -\alpha & 2\alpha + V_{N_x}
\end{array}\right)
\end{equation}

Since we are trying to diagonalize a matrix that is mostly zeros, we can use a special tridiagonal method to get eigenvalues/vectors. The conventional normalization can be done by dividing by $\delta_x$.

```python
def tise_1d(V,X,ħ=ħ,m=m,nvmax=-1):
    Nx = len(X)
    δx = np.abs(X[1]-X[0])
    α  = 0.5*ħ**2/m/δx**2
    
    diag    = V + 2.0*α
    subdiag = -α*np.ones(Nx-1)
    
    if (nvmax<0):
        E,Ψ = scipy.linalg.eigh_tridiagonal(diag,subdiag)
    else:
        E,Ψ = scipy.linalg.eigh_tridiagonal(diag,subdiag,
                                            select='i',
                                            select_range=(0,nvmax))
    
    Ψ /= np.sqrt(δx) # conventional normalization
    
    return E,Ψ
```

```python
# compute at t=0 since time-independent
%time E,Ψ = tise_1d(V[:,0],X)
```

```python
E_3 = np.full(Nx, 3*np.sqrt(3)*ħ*np.sqrt(κ/m) )
E_2 = np.full(Nx, (3/2)*ħ*np.sqrt(κ/m) )
plt.plot(X,V[:,0])

plt.plot(X,3*E_2,color='black',label="$3E^{(2)}(μ')$", linestyle='dashed')
plt.plot(X,E_3,color='blue',label="$E^{(3)}$", linestyle='dashed')
plt.ylim(0,4)
for iv in range(10):
    # the Ψ term in the second argument produces the wave in the figure
    # the coefficient determines the amplitude of the wave (purely for display purposes)
    plt.plot(X,E[iv] + 0.15*Ψ[:,iv], label=('n=1' if iv == 1 else ''))
plt.legend()
```

In the above figure, we can see the inequality $E^{(3)} \geq 3E^{(2)}$ in effect. The actual ground state ($n=1$) energy highlighted green is close to the predicted energy $E^{(3)}$ highlighted blue. Of course the green line accounts for the center of mass term as well, which is most likely where the difference lies. The lower bound (highlighted black) holds true and is below both the predicted and actual ground state energy value.
